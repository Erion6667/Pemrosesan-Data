```python
import pandas as pd

data = pd.ExcelFile("Hasil_Survei.xlsx")
df1 = pd.read_excel(data, "kinerja")
df2 = pd.read_excel(data, "harapan")
df1
```

```python
df2
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
importance = []
for col in df1.columns:
    column_mean = df1[col].mean()
    importance.append(column_mean)

# Calculate column means for df2
performance = []
for col in df2.columns:
    column_mean = df2[col].mean()
    performance.append(column_mean)
attribute = []
for i in range(1,16):
    attribute.append(i)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# membuat data frame
df = pd.DataFrame({
    'Attribute': attribute,
    'Importance': importance,
    'Performance': performance
})
#(analisis: disini kita menentukan titik dari atribut yang ada)

# membuat scatter plot
fig, ax = plt.subplots()
ax.scatter(df['Importance'], df['Performance'])
#(analisis: disini kita membuat plot dengan x-axis Importance dan y-axis Performance)

# menambahkan label untuk setiap data point
for i, txt in enumerate(df['Attribute']):
    ax.annotate(txt, (df['Importance'][i], df['Performance'][i]))
#(analisis: disini kita menambahkan kata atribut disebelah titik yang sudah ditentukan sebelumnya)
    
# menambahkan garis untuk menunjukkan nilai rata-rata Importance dan Performance
ax.axvline(df['Importance'].mean(), color='gray', linestyle='-')
ax.axhline(df['Performance'].mean(), color='gray', linestyle='-')
#(analisis: Disini kita menambahkan garis yang menunjukan nilai rata-rata dari Importance dan Performance)

# menambahkan label sumbu
ax.set_xlabel('performance')
ax.set_ylabel('Importance')
#(analisis: Disini kita menambahkan kata Importance dan Performance di samping plot)

# menambahkan judul plot
plt.title('Importance Performance Analysis')
#(analisis: Disini kita menambahkan judul di atas plot)

# menampilkan plot
plt.show()
#(analisis: Disini kita menampilkan plot)
```
